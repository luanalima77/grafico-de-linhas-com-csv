# Gráfico de linhas a partir de CSV

## Importação das bibliotecas

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

## Leitura do CSV

```python
dados = pd.read_csv("altura_mes.csv")
```

## Criando o gráfico de linhas

```python
plt.figure(figsize=(8,5))
plt.plot(dados["mes"], dados["altura"], marker="o", linewidth = 2)

plt.title("Altura ao longo do ano")
plt.xlabel("Mês")
plt.ylabel("Altura (em metros)")

plt.show()
```
